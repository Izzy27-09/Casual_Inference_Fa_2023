```python
brothers=['Gavin','Jay','Paul','Ro','Israel','Arjun','Jonah','Eric','Mahdi','Jackson','Cisco','Admir','Ethan','Max','Aaron']
```

```python
import random
```

```python
chores=['Dishes','Clean dinning room','Put away food','Living room','Empty Fridge/Freezer']
```

```python
days=['Mon','Tues','Wed','Thurs','Fri','Sat','Sun']
brother_c=brothers.copy()
for i in range(len(days)):
    print("\u0332".join(days[i]))
    if i !=6:
        for t in range(len(chores)-1):
            print('\033[1m'+chores[t]+'\033[0m')
            for p in range(2):
                pick=random.choice(brothers)
                print(pick)
                pick_place=brothers.index(pick)
                brothers.pop(pick_place) 
    else:
        for t in range(len(chores)):
            print('\033[1m'+chores[t]+'\033[0m')
            for p in range(2):
                pick=random.choice(brothers)
                print(pick)
                pick_place=brothers.index(pick)
                brothers.pop(pick_place) 
        
    print('\n')
    brothers=brother_c.copy()
    
```
