```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(1,5,0.1)
```

```python
tan_data=np.tan(x)

x_cub_data=np.power(x,3)
x_sqrt=np.sqrt(x)
```

```python
t=0
legends=['tan(x)','x^3','sqrt(3)']
while t<3:
    if t==0:
        plt.plot(x,tan_data,marker='x')
    if t==1:
        plt.plot(x,x_cub_data,marker='o',color='orange')
    if t==2:
        plt.plot(x,x_sqrt,color='green',linestyle='',marker='*')
    t+=1
plt.legend(legends)
plt.xlabel('X')
plt.ylabel('Y')
plt.savefig('ftb9-hw3.jpg')
```
